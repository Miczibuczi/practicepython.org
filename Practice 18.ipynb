```python
# Create a program that will play the “cows and bulls” game with the user. The game works like this:
# Randomly generate a 4-digit number. Ask the user to guess a 4-digit number. For every digit that the user guessed correctly in the correct place, they have a “cow”. For every digit the user guessed correctly in the wrong place is a “bull.” Every time the user makes a guess, tell them how many “cows” and “bulls” they have. Once the user guesses the correct number, the game is over. Keep track of the number of guesses the user makes throughout the game and tell the user at the end.
```

```python
import random
number = random.randint(1000,10000)

number_of_guesses = 1
guess = 0
def cow_bull_count():
    a, b, c, d = 0, 0, 0, 0
    cow = 0
    bull = 0
    string_guess = str(guess)
    a = string_guess[0]
    b = string_guess[1]
    c = string_guess[2]
    d = string_guess[3]
    string_number = str(number)
    A = string_number[0]
    B = string_number[1]
    C = string_number[2]
    D = string_number[3]
    if a == A:
        cow += 1
        a = None
        A = 10
    if b == B:
        cow += 1
        b = None
        B = 10
    if c == C:
        cow += 1
        c = None
        C = 10
    if d == D:
        cow += 1
        d = None
        D = 10
    list = [A,B,C,D]
    if a in list:
        bull += 1
        list.remove(a)
    if b in list:
        bull += 1
        list.remove(b)
    if c in list:
        bull += 1
        list.remove(c)
    if d in list:
        bull += 1
        list.remove(d)
    if cow == 1 and bull == 1:
        print(cow, "cow,", bull, "bull")
    elif cow == 1 and bull != 1:
        print(cow, "cow", bull, "bulls")
    elif cow != 1 and bull == 1:
        print(cow, "cows", bull, "bull")
    elif cow != 1 and bull != 1:
        print(cow, "cows", bull, "bulls")
    else:
        print("Some error occured")
    
        

import re
print("Welcome to the Cows and Bulls Game! Please enter a 4 digit number\n")
repeat = True
while repeat:
    guess = input("Enter a number\n")
    if re.match("\D", guess):
        print("Please enter a 4 digit number without any non-decimal characters")
    elif int(guess)<1000 or int(guess)>9999:
        print("Please enter a 4 digit number(a number between 1000 and 9999)")
    else:
        if int(guess) == number:
            print("Congratulation, you won! Number of guesses:", number_of_guesses)
            repeat = False
        else:
            cow_bull_count()
            number_of_guesses += 1
```
