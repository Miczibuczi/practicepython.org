```python
list1 = [0,0,0]
list2 = [0,0,0]
list3 = [0,0,0]
# 0 - no one put a token in that space
# 1 - player 1 put a token i that space
# 20 - player 2 put a token i that space



list_0 = [x, x1, x1]
list_1 = [y, y1, y1]
list_2 = [y, y1, y1]
list_3 = [y, y1, y1]


Player1_wins = 0
Player2_wins = 0
Play = True
Play_new_game = True

def draw_board():
    global list_drawing
    for i in range(3):
        for j in list_0:
            print(j, end = " ")
        print("")
        for k in list_drawing[i]:
            print (k, end = " ")
        print("")
    print(x, x1, x1)


def check_win():
    global Play
    global Player1_wins
    global Player2_wins
    # kolumns:
    sum1 = list1[0]+list2[0]+list3[0]
    sum2 = list1[1]+list2[1]+list3[1]
    sum3 = list1[2]+list2[2]+list3[2]

    # rows:
    sum4 = list1[0]+list1[1]+list1[2]
    sum5 = list2[0]+list2[1]+list2[2]
    sum6 = list3[0]+list3[1]+list3[2]

    # diagonals:
    sum7 = list1[0]+list2[1]+list3[2]
    sum8 = list1[2]+list2[1]+list3[0]

    # win condition:
    if sum1 == 3 or sum2 == 3 or sum3 == 3 or sum4 == 3 or sum5 == 3 or sum6 == 3 or sum7 == 3 or sum8 == 3:
        print("Player 1 wins")
        Player1_wins += 1
        Play = False
        return Play
    elif sum1 == 60 or sum2 == 60 or sum3 == 60 or sum4 == 60 or sum5 == 60 or sum6 == 60 or sum7 == 60 or sum8 == 60:
        print("Player 2 wins")
        Player2_wins += 1
        Play = False
        return Play


def game():
    global Play
    global list1
    global list2
    global list3
    global list_1
    global list_2
    global list_3
    global list_drawing
    turn = "Player 1"
    list1 = [0,0,0]
    list2 = [0,0,0]
    list3 = [0,0,0]
    x = " ---"
    x1 = "---"
    y = "|   |"
    y1 = "  |"
    list_0 = [x, x1, x1]
    list_1 = [y, y1, y1]
    list_2 = [y, y1, y1]
    list_3 = [y, y1, y1]
    list_drawing = [list_1,
                list_2,
                list_3]
    game_lists = [list1, 
                  list2,
                  list3]
    
    possible_moves = ["1,1", "1,2", "1,3", "2,1", "2,2", "2,3", "3,1", "3,2", "3,3"]
    
    y_X = "| X |"
    y1_X = "X |"
    y_O = "| O |"
    y1_O = "O |"
    
    while Play:
        if len(possible_moves) == 0:
            print("End of the game. It's a draw")
            break

            
        if turn == "Player 1":
            print("\nPlayer 1 turn. Please enter coordinates where do you want to place an X.\nPossible moves are:", end = "   ")
            for i in range(len(possible_moves)):
                print (possible_moves[i], end = "   ")
            move = input("\n")
            if move not in possible_moves:
                continue
            else:
                possible_moves.remove(move)
                game_lists[int(move[0])-1][int(move[2])-1] = 1
                if list_drawing[int(move[0])-1][int(move[2])-1] == y:
                    list_drawing[int(move[0])-1][int(move[2])-1] = y_X
                elif list_drawing[int(move[0])-1][int(move[2])-1] == y1:
                    list_drawing[int(move[0])-1][int(move[2])-1] = y1_X
                turn = "Player 2"
                draw_board()
                check_win()
        if Play == False:
            break
        if len(possible_moves) == 0:
            print("End of the game. Is's a draw")
            break
            
        if turn == "Player 2":
            print("\nPlayer 2 turn. Please enter coordinates where do you want to place an O.\nPossible moves are:", end = "   ")
            for i in range(len(possible_moves)):
                print (possible_moves[i], end = "   ")
            move = input("\n")
            if move not in possible_moves:
                continue
            else:
                possible_moves.remove(move)
                game_lists[int(move[0])-1][int(move[2])-1] = 20
                if list_drawing[int(move[0])-1][int(move[2])-1] == y:
                    list_drawing[int(move[0])-1][int(move[2])-1] = y_O
                elif list_drawing[int(move[0])-1][int(move[2])-1] == y1:
                    list_drawing[int(move[0])-1][int(move[2])-1] = y1_O
                turn = "Player 1"
                draw_board()
                check_win()

while Play_new_game:
    game()
    print("Player 1   :   Player 2")
    print("  ", Player1_wins, "            ", Player2_wins)
    new_game = input('Do you want to play again? Write "Yes" or "No"\n')
    if new_game.lower() == "yes":
        Play = True
        Play_new_game = True
    elif new_game.lower() == "no":
        Play_new_game = False
    else:
        while True:
            new_game = input('Do you want to play again? Write "Yes" or "No"\n')
            if new_game.lower() == "yes":
                Play_new_game = True
                Play = True
                break
            elif new_game.lower() == "no":
                Play_new_game = False
                break
```
