```python
# Make a two-player Rock-Paper-Scissors game. (Hint: Ask for player plays (using input), compare them, print out a message of congratulations to the winner, and ask if the players want to start a new game)
```

```python
def game():
    A = str(input("What do you chose, Rock, Paper or Scissors?\n"))
    B = str(input("What do you chose, Rock, Paper or Scissors?\n"))

    if A == B:
        print("It's a draw")
    elif A == "Rock" and B == "Scissors":
        print("Player 1 wins")
    elif A == "Rock" and B == "Paper":
        print("Player 2 wins")
    elif A == "Paper" and B == "Rock":
        print("Player 1 wins")
    elif A == "Paper" and B == "Scissors":
        print("Player 2 wins")
    elif A == "Scissors" and B == "Rock":
        print("Player 2 wins")
    elif A == "Scissors" and B == "Paper":
        print("Player 1 wins")
    else:
        print("You can choose Rock, Paper or Scissors")
        
def check():
    x = str(input("Do you want to play a new game? Write Yes or No\n"))
    if x == "Yes" or x == "yes":
        game()
        check()
    elif x == "No" or x == "no":
        pass
    else:
        check()
        
game()
check()
```
